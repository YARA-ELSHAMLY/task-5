```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread("C:/Users/moree/Downloads/task.jpeg")

gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, im3 = cv2.threshold(gray_img, 127, 255, cv2.THRESH_TRUNC)

imgBB = cv2.medianBlur(im3, )
imgBlur = cv2.bilateralFilter(imgBB, 9, 75, 75)
imgB = cv2.GaussianBlur(imgBlur, (5, 5), 0)
imgBBB=cv2.blur(imgB,(5,8))




brightness=20
contrast=1.7
iamge=cv2.addWeighted(imgB,contrast,np.zeros(imgB.shape,imgB.dtype),0,brightness)
iamge2=cv2.addWeighted(imgBlur,contrast,np.zeros(imgBlur.shape,imgBlur.dtype),0,brightness)



kernal=np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])
sharp_img=cv2.filter2D(iamge,-1,kernal)
sharp_img2=cv2.filter2D(iamge2,-1,kernal)
img2=cv2.cvtColor(sharp_img,cv2.COLOR_GRAY2BGR)




plt.figure(figsize=(10, 10))

plt.subplot(2, 3, 2), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)), plt.title('Original Image')
plt.subplot(2, 3, 3), plt.imshow(img2, cmap='gray'), plt.title('OUTPUT')


plt.show()
```
